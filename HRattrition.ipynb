Now we will delve deeper into the dataset and attempt to identify the reasoning behind organization attrition. Generally, employees leave their organization due to several reasons: Salary issues, difficulty working with superiors, bad work life balance, unsatisfactory with their job or poor workplace culture.


```python
ax1 = sns.countplot(x = 'Gender', hue = "Attrition", data = dataset2)
plt.title("Attrition rate between Gender")
for label in ax1.containers:
    labels = [f'{h/dataset2.Attrition.count()*100:0.1f}%' if (h := v.get_height()) > 0 else '' for v in label]
    ax1.bar_label(label, labels=labels, label_type='edge')
```

```python
ax2 = sns.countplot(y = 'JobRole', hue = 'Attrition', data=dataset2)
plt.title("Attrition rate between Job Role")
for label in ax2.containers:
    labels = [f'{h/dataset2.Attrition.count()*100:0.1f}%' if (h := v.get_width()) > 0 else '' for v in label]
    ax2.bar_label(label, labels=labels, label_type='edge')
```

```python
ax3 = sns.countplot(y = 'JobRole', hue = 'WorkLifeBalance', data=dataset2)
plt.title("Work Life Balance in each role")
for label in ax3.containers:
    labels = [f'{h/dataset2.WorkLifeBalance.count()*100:0.1f}%' if (h := v.get_width()) > 0 else '' for v in label]
    ax3.bar_label(label, labels=labels, label_type='edge')
```

```python
ax7 = sns.histplot(x = 'MonthlyIncome', hue = 'Attrition', multiple= "dodge", bins= 30, data=dataset2)
plt.title("Histogram for Monthly income")
```

The organization attrition rate is 16.1% which means that 83.9% of the employees did not choose to leave the organization. Males have the highest attrition rate at 10.2%. Comparing education level, it would seems that bachelor employees have higher attrition rate compared to other education levels.

Attrition between Job Role : Sales Representive, Lab Technician and Human Resources have the highest attrition rate compared to the other roles available within the organization. It is concerning and should be looked into why the attrition rate in these roles are significantly higher.
Looking at the work life balance, it does not appear that work life balance is a major factor as employees are generally happy with the work life balance.

Job and environment satisfaction could have played a part in the organization attrition rate as the number for low satisfaction are of a considerable amount (around 19.7% and 18.7% of the employee finds job and environment satisfaction to be low respectively.)

From the salary histogram, we can deduce that employees with lower monthly income (<5k) do not tend to stay in the company whereas employees with higher monthly income (>15k) are happy with their salary.

In conclusion, organization attrition rate can be reduced by improving the job and workplace culture within the organization. Ensuring that employees in the organization have benefits to motivate them and know that their well-being is being taken care of.
